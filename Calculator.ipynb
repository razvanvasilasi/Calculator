```python
# Imported clear_output() from IPython.display (The project was written in Jupyter Notebook)
from IPython.display import clear_output

# First, for each operation, a function was created wich later will be called.
# Every function was created for operations between two digits.

def Add(x,y):
    return x+y

def Subtract(x,y):
    return x-y

def Multiply(x,y):
    return x*y
    
def Divide(x,y):
    return x/y

def Percentace(x,y):
    return (x/100)*y

#Function for a nice calculator design, wich shows the result after each operation

def calculator_display(result):
    print('-------------------')
    print(f'{result}')
    print('-------------------')
    print(' + | - | * | / | % ')
    print('-------------------')
    print('|  7  |  8  |  9  |')
    print('-------------------')
    print('|  4  |  5  |  6  |')
    print('-------------------')
    print('|  1  |  2  |  3  |')
    print('-------------------')

#Continue or Stop Function

def again():
    choice=''
    while choice != 'Y' or choice != 'N':
        choice = input('Continue operating? (Y/N):').upper()
        return choice

```
